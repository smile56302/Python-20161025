# 選擇權payoff 繪圖

```python
%matplotlib notebook  
#%matplotlib inline   
#inline = 靜態  notebook為互動式
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
K = 9100
Interval = 500
Premium_Call = 179
Premium_Put = 185
ST = np.arange(K - Interval, K + Interval) # Vectorization
```

```python
# Payoff計算
Payoff_LongCall = np.maximum(ST - K, 0) - Premium_Call
Payoff_ShortCall = -Payoff_LongCall
Payoff_LongPut = np.maximum(K -ST , 0) - Premium_Put
Payoff_ShortPut = -Payoff_LongPut
```

```python
#Call 
plt.subplot(211)
plt.plot(ST, Payoff_LongCall, label = 'LongCall')
plt.plot(ST, Payoff_ShortCall, label= 'ShortCall')
plt.legend() # 呼叫圖例
plt.text(8800, 300, 'k=9100')  #http://matplotlib.org/users/pyplot_tutorial.html
#plt.xlabel('K', fontsize = 10, color = 'yellow')
plt.ylabel('Payoff', fontsize = 20, color = 'yellow')
plt.title('Payoff_Call', fontsize = 30 )
plt.axhline(0, color = 'black', ls = '--') #ls = linestyle, http://liam0205.me/2014/09/11/matplotlib-tutorial-zh-cn/
plt.axvline(K, color = 'black', ls = '--')
plt.axvline(K + Premium_Call, color = 'black', ls = '--')   # -- = dashed
#plt.ylabel('some numbers')
#plt.show()
#plt.savefig("filename.png",dpi=300,format="png")#將圖存成圖檔 #http://onionys.blogspot.tw/2012/01/python-matplotlib.html
#plt.subplot(212) 這個指令就是拿來畫多張圖用的指令。 212 = 2(列) * 1(行) 的第2張圖, 235 = 2*3 第5張圖
#plt.xticks(10,20,30) 來設定x軸刻度，plt.yticks(10,20,30)來設定y軸刻度


#PUT
plt.subplot(212)
plt.plot(ST, Payoff_LongPut, label = 'LongPut')
plt.plot(ST, Payoff_ShortPut, label = 'ShortPut')
plt.legend()
plt.text(9200, 300, 'K=9100')
plt.title('Payoff_Put', fontsize = 30)
plt.axhline(0, color = 'k', ls='--')
plt.axvline(K, color = 'k', ls='--')
plt.axvline(K - Premium_Put, color = 'k', ls='--')
```
